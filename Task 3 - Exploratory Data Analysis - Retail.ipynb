```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['State'].unique()
```

```python
plt.figure(figsize=(15,5))
ed=sns.countplot(x='State',data=df)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.show()
```

```python
mean_sales_state = df.groupby(['State'])['Sales'].mean().reset_index()
plt.figure(figsize=(12,5))
ed=sns.barplot(x='State',y='Sales',data=mean_sales_state)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.ylabel('Mean Sales',fontsize=15)
plt.xlabel('State',fontsize=15)
plt.show()
```

```python
mean_sales_state = df.groupby(['State'])['Profit'].mean().reset_index()
plt.figure(figsize=(12,5))
ed=sns.barplot(x='State',y='Profit',data=mean_sales_state)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.ylabel('Mean Profit',fontsize=15)
plt.xlabel('State',fontsize=15)
plt.show()
```
